### Top 10 countries with production count

```python
top_10_countries=titles_df.groupby('production_countries').id.count().sort_values(ascending =False).head(10)
```

```python
top_10_countries
```

```python
top_10_countries.plot(kind='bar',figsize=(8,4),color='purple')
plt.xlabel('Countries')
plt.ylabel('Production Count')
plt.title('Top 10 Countires with highest production');
```

## Question and Answer

### Q-1) Who is the best director? 

```python
credits_df.head(2)
```

```python
## Merging both tables to get details of actors and directors.
new_data=titles_df.merge(credits_df,how='outer',on='id')
```

```python
new_data.head(2)
```

```python
top_directors=new_data[new_data.role=='DIRECTOR'].sort_values(['tmdb_score','tmdb_popularity'],ascending=False).head(10)[['name','tmdb_score','tmdb_popularity']]
top_directors
```

```python
top_directors.plot(kind='bar',x='name',y='tmdb_popularity',color='pink')
plt.title('Top Directors');
```

Tooniverse is the best director.

### Q-2) Who is the best actor? 

```python
top_actors=new_data[new_data.role=='ACTOR'].sort_values(['tmdb_score','tmdb_popularity'],ascending=False).head(10)[['name','tmdb_score','tmdb_popularity']]
top_actors
```

```python
top_actors.plot(kind='bar',x='name',y='tmdb_popularity',color='blue')
plt.title('Top Actors');
```

This all are the best actors based on ratings

## Q-3) which the longest movie and its duration?

```python
duration=titles_df.sort_values('runtime',ascending=False).head(5)[['title','runtime']]
```

```python
duration
```

```python
duration.plot(kind='bar',x='title')
plt.xlabel('Name')
plt.ylabel('Time duration')
plt.title('Longest Movie');
```

The School of Mischief is the longest movie of duration	251 minutes.

## Q-4) which are the lowest 5 releases based on their ratings? 

```python
flop_release=titles_df.sort_values(['tmdb_score','tmdb_popularity']).head(5)[['title','tmdb_score','tmdb_popularity']]
flop_release
```

```python
flop_release.plot(kind='bar',x='title')
plt.xlabel('Name')
plt.ylabel('Ratings')
plt.title('5 Lowest Releases');
```
